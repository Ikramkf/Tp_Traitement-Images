```python
#importation des bibilotheques necessaires
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from skimage import data 
from skimage import color 
```

```python
#Importation des images 
coffee_image=data.coffee()
coins_image=data.coins()
```

```python
#1-Visualiser les images en utilisant la fonction show_image
def show_image(image, title='Coffee', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
#Thresholding 
#Import the threshold function 

from skimage.filters import threshold_otsu
```

```python
chess_pieces_image = plt.imread('./bw.jpg')
```

```python
show_image(chess_pieces_image,'Chess pieces')
```

```python
#Turn the image to grayscale 
chess_pieces_image_gray = color.rgb2gray(chess_pieces_image)
```

```python
#Obtain the optimal threshold value of the image 
thresh = threshold_otsu(chess_pieces_image_gray)
```

```python
#Apply thresholding to the image
binary = chess_pieces_image_gray > thresh 
```

```python
show_image(binary , "Binary image ")
```

```python
#When the background isn't obvious 
page_image = plt.imread('./text_page.png')
```

```python
show_image(page_image , 'Text ')
```

```python
from skimage.filters import threshold_local
```

```python
#Set the block size to 35 
block_size = 35
```

```python
#Obtain the optimal local thresholding 
local_thresh = threshold_local(page_image, block_size , offset=0.1)
```

```python
#Obtain the binary image by applying local thresholding 
binary_local = page_image > local_thresh
```

```python
show_image(binary_local , 'Local thresholding')
```
