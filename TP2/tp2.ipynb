```python
#importation des bibilotheques necessaires
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
import skimage 
import skimage.filters
from skimage import data
```

```python
#charger l'image cell
cell=data.cell()
```

```python
#1-Visualiser les images en utilisant la fonction show_image
def show_image(image, title='cell', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
show_image(cell,'cell_image' )
```

```python
#Multiplication par une constante inferieure a 1
cell_f= skimage.img_as_float(cell)
a=.5
Img_cte_inf=a*cell_f
plt.imshow(Img_cte_inf,'gray',vmin=0,vmax=1)
plt.show()
```

```python
#multiplication par une constante superieur à 1
b=1.5
Img_cte_sup=b*cell_f
plt.imshow(Img_cte_sup,'gray',vmin=0,vmax=1)
plt.show()
```

```python
#Addition avec une valeur inferieure a 1
c=0.5
Img_cte_add_inf=cell_f+c
plt.imshow(Img_cte_add_inf , 'gray')
plt.show()
```

```python
#Addition avec une valeur superieure a 1
c=3
Img_cte_add_sup=cell_f+c
plt.imshow(Img_cte_add_sup , 'gray')
plt.show()
```

```python
#Divison par une valeur inferieure à 1
d=0.5
Img_cte_div_inf=cell_f//d
plt.imshow(Img_cte_div_inf,'gray')
plt.show()
```

```python
#Divison par une valeur superieure à 1
d=2
Img_cte_div_sup=cell_f//d
plt.imshow(Img_cte_div_sup,'gray')
plt.show()
```

```python
#Multiplication image par une image 
Img_multipié=cell*cell_f
plt.imshow(Img_multipié,'gray')
plt.show()
```

```python
#Addition image avec une autre image
Img_additionée=cell+cell_f
plt.imshow(Img_additionée,'gray')
plt.show()
```

```python
#Divison image sur une autre image 
Img_divisée=cell//cell_f
plt.imshow(Img_divisée,'gray')
plt.show()
```

```python
#Filtres linéaires
import skimage.filters
```

```python
#Visualiser l'image
camera=skimage.data.camera()
show_image(camera,'Camera')
```

```python
plt.imshow(camera,'gray')
plt.show()
```

```python
from scipy.ndimage.filters import  uniform_filter
I_moy=uniform_filter(camera,size=(150,150))
plt.imshow(I_moy,'gray')
plt.show()
```

```python
I_median=skimage.filters.median(camera)
plt.imshow(I_median,'gray')
plt.show()
```

```python
I_gauss=skimage.filters.gaussian(camera,15)
plt.imshow(I_gauss,'gray')
plt.show()
```

```python
I_sobel=skimage.filters.sobel(camera)
#on calcule le gradient avec cette fonction
plt.imshow(I_sobel,'gray')
plt.show()
```

```python
I_shaar=skimage.filters.scharr(camera)
plt.imshow(I_shaar,'gray')
plt.show()
#filtres de gradiant , laissent passer les contours 
```

```python
I_hessian=skimage.filters.hessian(camera)
plt.imshow(I_hessian,'gray')
plt.show()
```
